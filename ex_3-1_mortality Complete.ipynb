# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide.plot.line(y='15-19 Years', ylabel='Deaths per 100,000',legend=False)
```

```python
mortality_wide.plot.area(legend='reverse')
```

```python
mortality_wide.query('1900').plot.bar()
```

```python
mortality_wide.query('2000').plot.bar(title='Child Mortality for the Year 2000', ylabel='Deaths per 100,000', xlabel='') 
```

```python
mortality_data.query('Year==2000').pivot(columns='AgeGroup', values='DeathRate').plot.bar(title='Child Mortality for the Year 2000', ylabel='Deaths per 100,000', xlabel='AgeGroup')
```

```python
mortality_data.plot.hist(y='DeathRate', bins=15) 
```

```python
mortality_data.query('Year == 1900')['DeathRate'].plot.density(title='Children Mortality for the Year 1900', grid=True, y='DeathRate')
```

```python
mortality_data.query('Year in [1900, 1925, 1950, 1975, 2000]').groupby('Year')['DeathRate'].sum().plot.pie(title='Total Child Mortality by Year', ylabel='')
```
