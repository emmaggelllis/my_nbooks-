## Fix spelling and capitalization problems in the data

```python
cars['brand'] = cars.apply(lambda x: x.CarName.split(' ')[0], axis=1)
cars['name'] = cars.apply(lambda x: ' '.join(x.CarName.split(' ')[1:]), axis=1)
```

```python
cars['brand'].unique()
```

```python
cars['brand'].str.strip().str.title().replace({ 'bmw':'BMW', 'maxda':'Mazda', 'porcshce':'Porsche', 'toyouta':'Toyota','vokswagen':'Volkswagen',  'vw':'Volkswagen'})
```

```python
cars['CarName'] = cars['brand'] + ' ' + cars['name']
cars.head()
```

## Rename and drop columns

```python
cars=cars.rename(columns={'CarName':'Brandname','car_ID':'Carid'})
cars.head()
```

```python
cars.drop(columns=['Carid','symboling'], inplace=True)
cars.head()
```
